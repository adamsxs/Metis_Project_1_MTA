```python
points = [point5, point6, point7]
def coords_max_min(points):
   '''
   Accepts a list of coordinate pairs.
   Returns a tuple with the maximum and minimum longitude and lattitudes
   (max_lat, min_lat, max_longs, min_longs)
   '''

   lats, longs = [], []

   for point in points:
       lats.append(point[0])
       longs.append(point[1])

   return (max(lats), min(lats), max(longs), min(longs))

coords_max_min(points)
```

```python
lat_max, lat_min, long_max, long_min = coords_max_min(points)
```
