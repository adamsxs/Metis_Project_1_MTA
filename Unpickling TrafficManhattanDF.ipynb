```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
with open("TrafficManhattanDF2018", 'rb') as picklefile: 
    TrafficManhattanDF2018 = pkl.load(picklefile)
    
with open("TrafficManhattanDF2017", 'rb') as picklefile: 
    TrafficManhattanDF2017 = pkl.load(picklefile)
    
with open("TrafficManhattanDF2016", 'rb') as picklefile: 
    TrafficManhattanDF2016 = pkl.load(picklefile)
    

    
TrafficManhattanDF = pd.concat([TrafficManhattanDF2016, TrafficManhattanDF2017, TrafficManhattanDF2018])
TrafficManhattanDF.groupby(['C/A', 'UNIT', 'STATION', 'DATE']).sum().sort_values(['DAILY_TRAFFIC'], ascending=False)





    
```
